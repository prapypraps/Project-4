```python
# Treating the columns to convert them into int type so that we have more variables to create model more accurate


# Removing the 'MP' text and converting the columns to integer type

df['Rear Camera'] = df['Rear Camera'].str.replace('MP', '').astype(int)
df['Front Camera'] = df['Front Camera'].str.replace('MP', '').astype(int)

# Verifing the changes by displaying the first few rows again

df[['Model', 'Rear Camera', 'Front Camera']].head()
```

```python
# Removing commas from the 'Prize' column and converting it to numeric values

df['Prize'] = df['Prize'].str.replace(',', '').astype(float)

# Verifing the changes by displaying the first few rows again

df[['Model', 'Prize']].head()
```

```python
# dropping unnecessary columns

df3 = df.drop(columns=['Unnamed: 0','Model', 'Colour'])

# model is removed as it is object type and model cant be included as a feature 

# here we removed color as its unique values are more than 50 % of the total valies
```

```python
# Treating the ouliers by defining a function whisper

def wisker (col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
# Treating the ouliers for better model accuracy 
# note- here we didnt use processor columns as it had no outliers
for i in ['Memory', 'RAM', 'Battery_', 'Rear Camera', 'Front Camera', 'AI Lens',
       'Mobile Height',  'Prize']:
    lw,uw=wisker(df3[i])
    df3[i]=np.where(df3[i]<lw,lw,df[i])
    df3[i]=np.where(df3[i]>uw,uw,df[i])
```

```python
# Checking whether outliers have been treated or now

for i in ['Memory', 'RAM', 'Battery_', 'Rear Camera', 'Front Camera', 'AI Lens',
       'Mobile Height', 'Processor_', 'Prize']: # checking wether it is resolved or now
    sns.boxplot(df3[i],color='darkblue')
    plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df4 = pd.DataFrame(df3)
```

```python
# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(df4)

# Convert the scaled data back to a DataFrame for easy handling
scaled_df = pd.DataFrame(scaled_data, columns=df4.columns)

# Show the scaled data
print(scaled_df)
```

```python
scaled_df=scaled_df.drop(244) # this particular row was deleted as it had a complex error which was not getting treated
```

```python
scaled_df
```

```python
scaled_df.columns
```

```python
# defining independent and dependent variables for prediction


X = scaled_df[['Memory', 'RAM', 'Battery_', 'Rear Camera', 'Front Camera', 'AI Lens',
       'Mobile Height', 'Processor_']]  # Independent variables
y = scaled_df['Prize']  # Dependent/target variable
```

```python
# importing train_test_split

from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,test_size=0.2,random_state=42)
```

```python
#importing linear regression 

from sklearn.linear_model import LinearRegression

# Initialize the linear regression model
model = LinearRegression()

```

```python
# Training the model using the training data

model.fit(X_train, y_train)
```

```python
# Making predictions on the test data

y_pred = model.predict(X_test)


y_pred
```

```python
pd.DataFrame({'Y_Test':y_test,'Y_Pred':y_pred})
```

```python
# checking the coefficient

pd.DataFrame(model.coef_,X.columns,columns=['Coefficients'])
```

```python
# Checking the accuracy of the model and knowing other score of the data
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_test,y_test).round(5)
```

```python
model.score(X_train,y_train).round(5)
```

```python
from sklearn import metrics
```

```python
MAE = metrics.mean_absolute_error(y_test,y_pred)
MSE = metrics.mean_squared_error(y_test,y_pred)
RMSE = np.sqrt(MSE)
```

```python
pd.DataFrame([MAE,MSE,RMSE],index=['MAE','MSE','RMSE'],columns=['Metrics'])
```

```python
# Plotting all features against Prize with regression lines
features = X_test.columns
plt.figure(figsize=(15, 20))

for i, feature in enumerate(features, 1):
    plt.subplot(3, 3, i)
    sns.regplot(x=X_test[feature], y=y_test, line_kws={"color": "red"}, scatter_kws={"alpha": 0.6})
    plt.xlabel(feature)
    plt.ylabel('Prize')
    plt.title(f'{feature} vs Prize')

plt.tight_layout()
plt.show()
```
